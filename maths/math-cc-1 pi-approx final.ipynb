## Leibniz Formula

The **Leibniz series** is an infinite sum whose value is $\pi$. It's terms are given as:

$$
\pi = 4\left(1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \frac{1}{9} + \cdots\right)
$$

We can use this series to numerically approximate the value of $\pi$, by evaluating the first $n$ terms of the Leibniz series.

**Question 1**: Define a function `Leibniz` which returns the *sum* of the first $n$ terms of the Leibniz series (including the factor of 4). Your function should take one input, $n$. You can check by hand (or with a calculator) if your function returns the correct answer for the first small values of $n$, e.g. `Leibniz`(2) should return 2.66666666..., and then check that your sum approaches $\pi$ for large values of $n$.

```python
def Leibniz(n):
    return sum[4*(-1)**(k)/(2k+1) for k in range(n)]
#After 'return', enter the object you want to be returned.


print(Leibniz(2))
#This runs Leibniz(2), and tells Python to print the result so we can see it. We don't need to use 'print' as Jupyter will print the last calculation, but it's good practice!
```

## Bonus: Monte Carlo Approximation

Another method of approximating $\pi$ is to use the **Monte Carlo** method. This involves generating points in a unit square, with corners at (0,0), (1,0), (1,1) and (0,1). We then check if this point also lies in the unit circle (only a quarter of which will be in the square), as shown in the picture below.

The proportion of points that lie in the unit circle should then approximately be the ratio of their areas, which in our case is $\frac{\pi}{4}$. Hence, to find an approximate value of $\pi$, we generate a large number $N$ of points, and count how many lie in the unit circle (say $U$). We should then have that $\frac{U}{N} \approx \frac{\pi}{4}$, which rearranges to give the approximation $\pi \approx \frac{4U}{N}$.
